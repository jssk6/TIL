### 라이브러리 간단 소개
- geopandas
    - geo 객체들을 pandas 데이터프레임으로 사용
- geoplot
    - geo 데이터 시각화
- osmnx
    - street networks 시각화, 알고리즘
- pysal
    - Python Spatial Analysis Library 
- contextily
    - Context geo-tiles in Python
- rasterio
    - geospatial raster 데이터 접근
- rasterstats
    - Python module for summarizing geospatial raster datasets based on vector geometries
